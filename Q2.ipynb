```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/(3*h)
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
## CS1
```

```python
img1 = cv2.imread('image1.jfif')
CS1 = -1 * img1 + 255
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("ORG")
plt.subplot(1,2,2)
plt.title("CS1")
plt.imshow(cv2.cvtColor(np.uint8(CS1), cv2.COLOR_BGR2RGB))
plt.show()
```

```python
## CS2
```

```python
row1 , col1 = img1.shape[:2]
```

```python
CS2 = img1.copy()
for i in range(0,row1):
    for j in range(0,col1):
        for k in range(0,3):
            if img1[i][j][k]>.2*255 and img1[i][j][k]<.55*255:
                CS2[i][j][k] = 0
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("ORG")
plt.subplot(1,2,2)
plt.title("CS2")
plt.imshow(cv2.cvtColor(np.uint8(CS2), cv2.COLOR_BGR2RGB))
plt.show()
```

```python
## CS3
```

```python
CS3 = img1.copy()
for i in range(0,row1):
    for j in range(0,col1):
        for k in range(0,3):
            if img1[i][j][k]>.55*255:
                CS3[i][j][k] = 255
            elif img1[i][j][k]<.4*255:
                CS3[i][j][k] = 0
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("ORG")
plt.subplot(1,2,2)
plt.title("CS3")
plt.imshow(cv2.cvtColor(np.uint8(CS3), cv2.COLOR_BGR2RGB))
plt.show()
```

```python
## CS4
```

```python
img1_org = img1 / 255
CS4 = img1_org.copy()
for i in range(0,row1):
    for j in range(0,col1):
        for k in range(0,3):
            if img1_org[i][j][k]<.2:
                CS4[i][j][k] *= 1.5
            elif img1_org[i][j][k]<.55:
                CS4[i][j][k] = .3
            else:
                CS4[i][j][k] = CS4[i][j][k] *(.7 / .45) - (.7/.45) + 1
CS4 *= 255
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("ORG")
plt.subplot(1,2,2)
plt.title("CS4")
plt.imshow(cv2.cvtColor(np.uint8(CS4), cv2.COLOR_BGR2RGB))
plt.show()
```
