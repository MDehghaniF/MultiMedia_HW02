```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/(3*h)
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
img2 = cv2.imread('trees.jpeg')
B2, G2, R2 = cv2.split(img2)
```

```python
histSize = 256
histRange = (0, 256) # the upper boundary is exclusive
```

```python
zeros = np.zeros(img2.shape[:2], dtype = "uint8")
r_hist = cv2.calcHist([img2], [2], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("ORG")
plt.imshow(cv2.cvtColor(cv2.merge([zeros, zeros, R2]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.xlim([0,256])
plt.plot(r_hist)
plt.title("R Histogram")
plt.show()
```

```python
zeros = np.zeros(img2.shape[:2], dtype = "uint8")
g_hist = cv2.calcHist([img2], [1], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("ORG")
plt.imshow(cv2.cvtColor(cv2.merge([zeros, G2, zeros]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("g Histogram")
plt.plot(g_hist)
plt.show()
```

```python
b_hist = cv2.calcHist([img2], [0], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("ORG")
plt.imshow(cv2.cvtColor(cv2.merge([B2, zeros , zeros]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("B Histogram")
plt.plot(b_hist)
plt.show()
```

```python
R2_rsz = cv2.resize(R2,None, fx = .25, fy = .25, interpolation = cv2.INTER_NEAREST)
G2_rsz = cv2.resize(G2,None, fx = .25, fy = .25, interpolation = cv2.INTER_NEAREST)
B2_rsz = cv2.resize(B2,None, fx = .25, fy = .25, interpolation = cv2.INTER_NEAREST)
```

```python
zeros = np.zeros(R2_rsz.shape[:2], dtype = "uint8")
img2_rsz = cv2.merge([B2_rsz , G2_rsz , R2_rsz])
r_hist = cv2.calcHist([img2_rsz], [2], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("ORG")
plt.imshow(cv2.cvtColor(cv2.merge([zeros, zeros, R2_rsz]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("R Resized Histogram")
plt.plot(r_hist)
plt.show()
```

```python
g_hist = cv2.calcHist([img2_rsz], [1], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("ORG")
plt.imshow(cv2.cvtColor(cv2.merge([zeros, G2_rsz , zeros]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("G Resized Histogram")
plt.plot(g_hist)
plt.show()
```

```python
b_hist = cv2.calcHist([img2_rsz], [0], None, [histSize], histRange)
plt.subplot(1,2,1)
plt.title("B resized Channel")
plt.imshow(cv2.cvtColor(cv2.merge([B2_rsz , zeros , zeros]), cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title("b Resized Histogram")
plt.plot(b_hist)
plt.show()
```
