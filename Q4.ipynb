```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/(h)
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
img_org = cv2.imread('AerialView.jpeg')
img = img_org.copy()
```

```python
bilateral = cv2.bilateralFilter(img,21,75,75)
imshow("Original",img_org)
imshow("bilateral",bilateral)

Gaussian = cv2.GaussianBlur(img, (21,21), 0)
imshow("Gaussian",Gaussian)

median = cv2.medianBlur(img, 21)
imshow("Median",median)
```

```python
bcanny = cv2.Canny(bilateral, 50, 120)
imshow('bilateral Canny', bcanny)

gcanny = cv2.Canny(Gaussian, 50, 120)
imshow('Gaussian Canny', gcanny)

mcanny = cv2.Canny(median, 50, 120)
imshow('Median Canny', mcanny)
```

```python
MSX = cv2.Sobel(median, cv2.CV_64F,1,0,3)
MSY = cv2.Sobel(median, cv2.CV_64F,0,1,3)
MASX = cv2.convertScaleAbs(MSX)
MASY = cv2.convertScaleAbs(MSY)
SMedian = cv2.addWeighted(MASX, 0.5,MASY, 0.5, 0)
imshow("Median Sobel",SMedian)
```
